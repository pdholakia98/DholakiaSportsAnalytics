```python
#hide
import pandas as pd
import altair as alt
from altair import datum
import numpy as np
```

```python
#hide
#Read NBA Rookies of the Year into a dataframe

ROY_df_URL = 'https://www.basketball-reference.com/awards/roy.html'

ROY_df = pd.read_html(ROY_df_URL)[0]

ROY_df = ROY_df.droplevel(0, axis =1)

ROY_df.drop_duplicates(inplace = True)

#base_safety_inbox = customer_safety_inbox.drop(columns = ["following","turn","distracted","forward","brake" , "Event URL"])

ROY_df = ROY_df.drop(columns = ["Voting" , "Lg"])

ROY_df.head()
```

```python
#hide
#Read 2021 stats into a dataframe

PerGamePlayerStats2021URL = "https://www.basketball-reference.com/leagues/NBA_2021_per_game.html"

PerGamePlayerStats2021 = pd.read_html(PerGamePlayerStats2021URL)[0]

PerGamePlayerStats2021.drop_duplicates(inplace = True)

PerGamePlayerStats2021.head()
```

```python
#hide
Lamelo_2021 = PerGamePlayerStats2021[(PerGamePlayerStats2021.Player == "LaMelo Ball")]

Lamelo_2021.head()
```

```python
#hide
Combined = Lamelo_2021.append(ROY_df)

Combined = Combined.reset_index()
```

```python
#hide
alt.Chart(Combined).mark_circle().encode(
    x = "MP:Q",
    y = "PTS:Q"
).transform_filter(
   alt.FieldEqualPredicate(field = "Player" , equal = "LaMelo Ball")
)
```

```python
#hide

selection = alt.selection_single();
  
chart1 = alt.Chart(Combined).mark_circle().add_selection(
    selection
).encode(
    x='MP:Q',
    y='PTS:Q',
    opacity=alt.condition(selection, alt.value(0.8), alt.value(0.1))
)

chart2 = alt.Chart(Combined).mark_circle().encode(
    x = "MP:Q",
    y = "PTS:Q",
    color = alt.value("black")
).transform_filter(
   alt.FieldEqualPredicate(field = "Player" , equal = "LaMelo Ball")
)

chart1 + chart2
```

```python
#hide_input

hover = alt.selection_single(
    on='mouseover',  # select on mouseover
    nearest=True,    # select nearest point to mouse cursor
    empty='none'     # empty selection should match nothing
)

click = alt.selection_multi(
    empty='none' # empty selection matches no points
)

# scatter plot encodings shared by all marks
plot = alt.Chart(Combined).mark_circle().encode(
    x='MP:Q',
    y='PTS:Q'
).properties(
    title = "Lamelo Ball Minutes and Points Per Game vs Previous Rookies of the Year"
)

plot2 = alt.Chart(Combined).mark_circle().encode(
    x = "MP:Q",
    y = "PTS:Q",
    color = alt.value("red")
).transform_filter(
   alt.FieldEqualPredicate(field = "Player" , equal = "LaMelo Ball")
)
    
    
# shared base for new layers
base = plot.transform_filter(
    hover | click # filter to points in either selection
)

# layer scatter plot points, halo annotations, and title labels
alt.layer(
    plot.add_selection(hover).add_selection(click),
    base.mark_point(size=100, stroke='firebrick', strokeWidth=1),
    base.mark_text(dx=4, dy=-8, align='right', stroke='white', strokeWidth=2).encode(text='Player:N'),
    base.mark_text(dx=4, dy=-8, align='right').encode(text='Player:N'),
    plot2,
    data=Combined
).properties(
    width=600,
    height=450
)
```

```python
#hide_input

hover = alt.selection_single(
    on='mouseover',  # select on mouseover
    nearest=True,    # select nearest point to mouse cursor
    empty='none'     # empty selection should match nothing
)

click = alt.selection_multi(
    empty='none' # empty selection matches no points
)

# scatter plot encodings shared by all marks
plot = alt.Chart(Combined).mark_circle().encode(
    x='FG%:Q',
    y='3P%:Q'
).properties(
    title = "Lamelo Ball FG% and 3P% Compared to Previous Rookies of the Year"
)
 
plot2 = alt.Chart(Combined).mark_circle().encode(
    x = "FG%:Q",
    y = "3P%:Q",
    color = alt.value("red")
).transform_filter(
   alt.FieldEqualPredicate(field = "Player" , equal = "LaMelo Ball")
)
    
# shared base for new layers
base = plot.transform_filter(
    hover | click # filter to points in either selection
)

# layer scatter plot points, halo annotations, and title labels
alt.layer(
    plot.add_selection(hover).add_selection(click),
    base.mark_point(size=100, stroke='firebrick', strokeWidth=1),
    base.mark_text(dx=4, dy=-8, align='right', stroke='white', strokeWidth=2).encode(text='Player:N'),
    base.mark_text(dx=4, dy=-8, align='right').encode(text='Player:N'),
    plot2,
    data=Combined
).properties(
    width=300,
    height=450
)
```

```python
colstoconvert = ["MP",
                 "FG",
                 "FGA",
                 "FG%",
                 "3P",
                 "3PA",
                 "3P%",
                 "2P",
                 "2PA",
                 "2P%",
                 "TRB",
                 "AST",
                 "STL",
                 "BLK",
                 "PTS"
                ]

Combined[colstoconvert] = Combined[colstoconvert].apply(pd.to_numeric, errors = "coerce")
```

```python
#hide

#create a rank for Points, assists, steals

Combined["Minutes Ranked"] = Combined["MP"].rank(ascending = False)

Combined["Points Ranked"] = Combined["PTS"].rank(ascending = False)

Combined["Assists Ranked"] = Combined["AST"].rank(ascending = False)

Combined["Steals Ranked"] = Combined["STL"].rank(ascending = False)

Combined["FG% Ranked"] = Combined["FG%"].rank(ascending = False)

Combined["3P% Ranked"] = Combined["3P%"].rank(ascending = False)

Combined.head()
```
